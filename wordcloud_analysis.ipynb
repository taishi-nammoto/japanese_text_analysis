```python
import MeCab
import numpy as np
import pandas as pd
from PIL import Image
from matplotlib import pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')
```

```python
karada_genki = pd.read_excel('article_data.xlsx', sheet_name='karada_genki')    
otona_beauty = pd.read_excel('article_data.xlsx', sheet_name='otona_beauty')   
kounenki_no_chie = pd.read_excel('article_data.xlsx', sheet_name='kounenki_no_chie')   
life = pd.read_excel('article_data.xlsx', sheet_name='life')   
odekake_joshigumi = pd.read_excel('article_data.xlsx', sheet_name='odekake_joshigumi')
food = pd.read_excel('article_data.xlsx', sheet_name='food')
future_planning = pd.read_excel('article_data.xlsx', sheet_name='future_planning')
```

```python
def clean_text(df):
    text=''
    for article in df:
        text += article
        
    tagger = MeCab.Tagger('')
    tagger.parse('') 
    node = tagger.parseToNode(text)
    word_list = []
    while node:
        word_type = node.feature.split(',')[0]
        if word_type in ["名詞"]:   
            word = node.surface
            word_list.append(word)
        node = node.next
    return " ".join(word_list)
```

```python
text = clean_text(karada_genki['content'])

mask = np.array(Image.open('health.png'))

wc = WordCloud(background_color="black", mask=mask, max_words=200, width=4000, height=2000, 
               font_path='/Library/Fonts/Arial Unicode.ttf',
                ).generate(text) # contour_width=2, contour_color='red'

plt.figure(figsize=(10,5))
image_colors = ImageColorGenerator(mask) 
plt.imshow(wc.recolor(color_func=image_colors), interpolation='bilinear', aspect="equal")
plt.axis('off')
plt.show()
wc.to_file('health_content.png')
```
