```python
import pandas as pd
from PIL import Image
import seaborn as sns
from matplotlib import pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')
import japanize_matplotlib
import re
import operator
```

```python
dictDaily = {'健康':4, '生活':3, '旅行':2, '美容':1}
#dictDaily = dict(sorted(dictDaily.items(), key=operator.itemgetter(1),reverse=True))
```

```python
sns.set(font='IPAexGothic')
sns.set_context("notebook", font_scale=1.1)
pie = plt.pie(dictDaily.values(), labels = dictDaily.keys(), labeldistance=0.7, counterclock=False, 
              startangle=-270, wedgeprops={'alpha':0.5}, textprops=dict(ha='center'))
plt.title('日間ランキング')
plt.savefig('日間ランキング'+'.png', dpi=300);
```
